## Tóm tắt lý thuyết : <br>
> Trong bài thực hành trước, các tính chất trong miền thời gian của tín hiệu và hệ thống rời rạc thời gian đã được khảo sát. Tuy nhiên, nhiều tính chất quan trọng của tín hiệu và hệ thống rời rạc thời gian chỉ được thể hiện khi biểu diễn chúng trong miền tần số.<br>

## A. Tín hiệu rời rạc thời gian trong miền tần số <br>
### 1. Biến đổi Fourier rời rạc thời gian (DTFT) <br>
Biến đổi Fourier rời rạc thời gian của tín hiệu x[n] và biến đổi ngược của nó được định nghĩa bởi phương trình 1 và 2: <br>
$$X[\Omega]=DTFT(x[n])=\sum_{n=-\infty}^{+\infty}x[n]e^{-j\Omega n} (1)$$
$$x[n]=IDTFT(X[\Omega])=\frac{1}{2\pi}\sum_{n=-\pi}^{\pi}X[\Omega]e^{j\Omega n}d \Omega(2)$$ 
$X(\Omega)$ có thể được biểu diễn dưới dạng biên độ và pha như ở phương trình 3: <br>
$$X[\Omega] = |X[\Omega]|e^{j.arg(X[\Omega])} (3)$$
Khi đó $X(\Omega)$ được gọi là phổ tín biên độ tín hiệu và $arg(X[\Omega])$ được gọi là phổ pha của tín hiệu x[n]
### 2. Biến đổi Fourier rời rạc
> Cho công thức biến đổi Fourier rời rạc (DFT) như sau: <br>
$$X(k)=\sum_{n=0}^{N-1} x(n)e^{\frac{-2j\pi kn}{N}}$$ 

Trong đó:
* N là số mẫu <br>
* n là mẫu hiện tại <br>
* $x_{n}$ là gí trị của tín hiệu tại thời gian n <br>
* k là tần số hiện tại <br>
* $X_{k}$ là kết quả của DFT <br>

Bài 1a: <br>
Đoạn chương trình sau đây định nghĩa hàm tính DTFT

````python
def dtft(x):
    L = len(x)
    w = np.arange(-2*np.pi, 2*np.pi, 4*np.pi/512) #Omega từ -2pi đến 2pi
    w = w.reshape(len(w),1) #Chuyển vị
    n = np.arange(L)
    e = np.exp(-1j *w*n) # Lưu  ý: ta ghi 1j để biểu thị j là phần ảo của số phức
    X = np.dot(e, x )#Nhân vô hướng
    return w, X
````

Bài 2a: <br>
> Cho công thức biến đổi Fourier rời rạc (DFT) như sau: <br>
$$X(k)=\sum_{n=0}^{N-1} x(n)e^{-j2 \pi k n/N}$$
Trong đó: <br>
* N là số mẫu <br>
* n là mẫu hiện tại <br>
* $x_n$ là gí trị của tín hiệu tại thời gian n <br>
* k là tần số hiện tại <br>
* $X_k$ là kết quả của DFT <br>
 
Đoạn chương trình sau định nghĩa một hàm để tính DFT <br>


````python
def dft(x):
  N = len(x)
  X = np.zeros(N, dtype = complex)
  for k in range(N):
    X[k] = 0
    for n in range(N):
      X[k] = X[k] + x[n]*np.exp(-2j*np.pi*k*n/N) 
  return X
````

Cách khác:

````python
def dft1(x):
    """
    Tính biến đổi Fourier 1 chiều với đầu vào x
    :param x: (array)
    """

    N = len(x) #Lấy chiều dài của x
    n = np.arange(N)
    k = n.reshape((N, 1)) #Chuyển n thành mảng N hàng 1 cột
    e = np.exp(-2j * np.pi * k * n / N)
    return np.dot(e, x) #Nhân vô hướng 2 ma trận
````

Bài 3a: <br>
Hàm sau đây thực hiện nhân chập vòng tròn của 2 tín hiệu x1 và x2

````python
def circonv(x1, x2):
  L1 = len(x1)
  L2 = len(x2)
  if L1 != L2:
    print('Lỗi! Hai tín hiệu không bằng nhau')
  y = np.zeros(L1)
  x2[1:] = x2[L2-1:0:-1]
  x2tr = x2
  for k in range(L1):
    sh = np.roll(x2tr, k)# Dịch vòng tròn
    h = x1*sh
    y[k] = sum(h)
  return y
````

Bài 3b: <br>
Chương trình sau đây minh họa tính chất phép nhân thường trong miền thời gian và phép nhân chập trong miền tần số

````python
#Sử dụng nhân chập vòng tròn
x = np.array([1, 2, 3, 4])
h = np.array([1, -1, 1, 0]) # Đảm bảo cùng độ dài để thực hiện nhân chập vòng tròn


# Nhân chập vòng tròn
y_circular = circonv(x.copy(), h.copy())

# Nhân thường trong miền tần số (dùng FFT)
X = np.fft.fft(x)
H = np.fft.fft(h)
Y_freq = X * H
y_freq = np.fft.ifft(Y_freq).real  # Lấy phần thực

# Vẽ biểu đồ so sánh
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.stem(y_circular, basefmt=" ")
plt.title("Circular Convolution")
plt.xlabel("n")
plt.ylabel("y[n]")

plt.subplot(1, 2, 2)
plt.stem(y_freq, basefmt=" ")
plt.title("Multiplication in Frequency Domain")
plt.xlabel("n")
plt.ylabel("y[n]")

plt.tight_layout()
plt.show()
````

````python
#Sử dụng hàm convolve

# Tín hiệu mẫu x[n] và h[n]
x = np.array([1, 2, 3, 4])
h = np.array([1, -1, 1])

# Nhân chập trong miền thời gian
y_time = np.convolve(x, h)

# Nhân thường trong miền tần số
N = len(x) + len(h) - 1  # Độ dài cần thiết sau khi nhân chập
X = np.fft.fft(x, N)
H = np.fft.fft(h, N)
Y_freq = X * H
y_freq = np.fft.ifft(Y_freq).real  # Lấy phần thực vì kết quả lý tưởng là thực

# Vẽ biểu đồ so sánh
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.stem(y_time)
plt.title('Nhân chập trong miền thời gian')
plt.xlabel('n')
plt.ylabel('y[n]')

plt.subplot(1, 2, 2)
plt.stem(y_freq)
plt.title('Nhân thường trong miền tần số (IFFT)')
plt.xlabel('n')
plt.ylabel('y[n]')

plt.tight_layout()
plt.show()
````

````python
# Define signal x[n]
x = np.array([0, 0, 0, 0, 0, 1, 1.5, 1.7, 2, 1.5, 1, 0.5, 0, 0, 0, 0, 0])

# Calculate DTFT from 0 to pi
w = np.linspace(0, np.pi, 512)
n = np.arange(len(x))
X_dtft = np.zeros(len(w), dtype=complex)
for i in range(len(w)):
    X_dtft[i] = np.sum(x * np.exp(-1j * w[i] * n))

# Calculate DFT for k=0 to 8
X_dft = dft(x)[0:9]

# Create subplot
plt.figure(figsize=(10, 8))

# Plot DTFT magnitude
plt.subplot(2, 1, 1)
plt.plot(w / np.pi, np.abs(X_dtft))
plt.title('Magnitude of DTFT')
plt.xlabel('$\Omega/\pi$')
plt.ylabel('|X($\Omega$)|')
plt.grid(True)

# Plot DFT magnitude
plt.subplot(2, 1, 2)
plt.stem(range(9), np.abs(X_dft))
plt.title('Magnitude of DFT')
plt.xlabel('k')
plt.ylabel('|X[k]|')
plt.grid(True)

plt.tight_layout()
plt.show()
#DTFT là phổ "thật", DFT là phổ "thực tế":
#DTFT (đường cong liền mạch): Cho ta thấy phổ tần số liên tục và đầy đủ của tín hiệu. Đây là cái chúng ta muốn biết về mặt lý thuyết.
#DFT (các que rời rạc): Là cái chúng ta có thể tính toán được trong thực tế. Nó chỉ cung cấp giá trị phổ tại một số điểm tần số hữu hạn.
#Sự tương quan trực quan:
#Điều cốt lõi cần thấy là mỗi que trong biểu đồ DFT nằm chính xác trên đường cong của biểu đồ DTFT.
#Điều này chứng tỏ DFT không phải là một phép biến đổi hoàn toàn khác, mà nó chỉ đơn giản là "nhìn" vào phổ DTFT tại các điểm tần số cách đều nhau.
````

$y(n) = x(n) – 1.9021x(n-1) + x(n-2) + 1.8523y(n-1) – 0.94833y(n-2) $ <br>
Cho biết đây là lọc gì?

Bài 6b: <br>
Vẽ lại đáp ứng vẽ lại đáp ứng bộ lọc theo Hz với chu kì lấy mẫu T = 1/100

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define filter coefficients
b = [1, -1.9021, 1]      # Hệ số cho tín hiệu vào x
a = [1, -1.8523, 0.94833] # Hệ số cho tín hiệu ra y

# Tần số lấy mẫu và tần số Nyquist
fs = 100  # Tần số lấy mẫu (sample rate), T = 1/100
nyquistRate = fs / 2

# Tính toán đáp ứng tần số
w, H = signal.freqz(b, a, worN=512, whole=False)

# Chuyển đổi từ rad/sample sang Hz
freq = w / (2 * np.pi) * fs

# Vẽ đáp ứng biên độ và đáp ứng pha
plt.figure(figsize=(10, 6))

# Đáp ứng biên độ
plt.subplot(2, 1, 1)
plt.plot(freq, 20 * np.log10(np.abs(H)))
plt.title('Đáp ứng biên độ')
plt.xlabel('Tần số (Hz)')
plt.ylabel('Biên độ (dB)')
plt.grid(True)
plt.axvline(nyquistRate, color='r', linestyle='--', label='Nyquist Freq') # Thêm dòng cho tần số Nyquist

# Đáp ứng pha
plt.subplot(2, 1, 2)
plt.plot(freq, np.angle(H))
plt.title('Đáp ứng pha')
plt.xlabel('Tần số (Hz)')
plt.ylabel('Pha (rad)')
plt.grid(True)
plt.axvline(nyquistRate, color='r', linestyle='--') # Thêm dòng cho tần số Nyquist

plt.tight_layout()
plt.show()
````
